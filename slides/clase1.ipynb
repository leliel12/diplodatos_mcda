## Teoría racional de la decisión 

Un problema de Decisión se presenta cuando, frente a un problema, existe más de un curso de acción posible.

**Sujetos intervinientes:**
    
- **Decisor**: es quien tiene el poder y responsabilidad de tomar la decisión y responde por sus consecuencias.
- **Analista**: es el encargado de estructurar el problema y ayudar al decisor a visualizarlo.

**Elementos conocidos**

- Conjunto de alternativas:&nbsp;$X= \left\lbrace x_1, x_2,..., x_n\right\rbrace$
- Conjunto de estados de la naturaleza:&nbsp;$Y= \left\lbrace y_1, y_2,..., y_m\right\rbrace$
- Matriz de las compensaciones: 

$$\begin{bmatrix}
     C(x_1,y_1)&C(x_1,y_2) & ...  & C(x_1,y_m) \\
     C(x_2,y_1)&C(x_2,y_2) & ...  & C(x_2,y_m) \\
     ... & ... & ... & ... \\
     C(x_n,y_1)&C(x_n,y_2) & ...  & C(x_n,y_m) \\
    \end{bmatrix}$$


## Para decidir racionalmente se aplica un marco normativo justificado en: 

- La eficacia de las personas para alcanzar sus objetivos (*Maximización*).
- Las decisiones óptimas aumentan las posibilidades de supervivencia en un entorno competitivo.
- Axiomas proporcionan una explicación aceptable del comportamiento de elección (*Atractivo intuitivo*). 





## Reglas del modelo normativo 
- Cancelación
- Transitividad

- Invarianza

## Propiedad 2: Transitividad 

Las utilidades están representadas a través de las funciones de decisión d(x).

$A \succ B$ $\longrightarrow d(A)> d(B)$

|          | $y_1$ | $y_2$ | $$d(x) = E_y C(x,y)$$ |
|:--------:|:-----:|:-----:|:-------------------:|
|   $x_1$  |   8   |   2   |         3,2         |
|   $x_2$  |   4   |   6   |         5,6         |
| $P(y_i)$ | 0,20  | 0,80  |                     |


 En este caso $d(x_1)< d(x_2)$
 $x_2$ es preferible a $X_1$


## Propiedad 3: Dominancia

Se da cuando una opción es mejor que otra en un estado de la naturaleza y al menos tan buena en todos los demás estados.

|       | $$y_1$$ | $$y_2$$ | $$y_3$$ |   |       | $$y_1$$ | $$y_2$$ | $$y_3$$ |
|:-----:|:-----:|:-----:|:-----:|   |:-----:|:-----:|:-----:|:-----:|
| $x_1$ |   5   |   8   |   7   |   | $x_1$ |   6   |   2   |   3   |
| $x_2$ |   3   |   2   |   1   |   | $x_2$ |   5   |   2   |   3   |


## Propiedad 4: Invarianza 

Las diferentes representaciones  (marco) del mismo problema de elección deben producir la misma preferencia. Es decir, la preferencia entre las opciones debe ser independiente de su descripción. 

|          | $y_1$ | $y_2$ | $$d(x) = E_y C(x,y)$$ |   |          | $y_1$ | $y_2$ | $$d(x) = E_y C(x,y)$$ |
|:--------:|:-----:|:-----:|:-------------------:|---|:--------:|:-----:|:-----:|:-------------------:|
|   $x_1$  |   5   |   8   |         6,8         |   |   $x_1$  |   8   |   5   |         6,8         |
|   $x_2$  |   6   |   2   |         3,6         |   |   $x_2$  |   2   |   6   |         3,6         |
| $P(y_i)$ | 0,40  | 0,60  |                     |   | $P(y_i)$ | 0,60  | 0,40  |                     |

## Problema del Modelo Normativo 

Las **desviaciones** del comportamiento real del modelo normativo son **demasiado amplias** como para ser ignoradas, **demasiado sistemáticas** para ser descartadas como error aleatorio, y **demasiado fundamentales** para ser acomodadas relajando el sistema normativo.

Distintos teóricos respondieron a las violaciones observadas de cancelación y transitividad al debilitar la teoría normativa a fin de mantener su estatus como modelo descriptivo. 

`Una teoría de la decisión racional no puede proporcionar una descripción adecuada del comportamiento de elección.`

## `skcriteria.preprocessing.weighters.StdWeighter`

- Utiliza el desvio estandar por criterio normalizado.
- Se altera en caso de invertir los criteriios ($1/Ci$) o en normalizaciones.

$$w_j = \frac{s_j}{\sum{s}}$$

Donde:
- $s_j$ Es el devio muestral de cada criterio $j$.
- $\sum{s}$ es la sum de todos los $s_j$.

```python
wght = weighters.StdWeighter()
wght
```

```python
wght.transform(dm)
```

## `skcriteria.preprocessing.weighters.EntropyWeighter`

- En este caso el criterio con mas "sorpresas" es el mas importante.
- Bastante robusto frente a ruidos y normalizaciones.

$$
    w_j = \frac{H_j}{\sum{H}}
$$

Donde:
- $H_j$ entropía del criterio $j$.
- $\sum{H}$ la suma de las entropías de todos los criterios.

```python
wght = weighters.EntropyWeighter()
wght
```

```python
wght.transform(dm)
```

## Technique for Order Preference by Similarity to Ideal Solution (TOPSIS)

- Se basa en el concepto de que es deseable que una alternativa determinada se ubique a la distancia más corta respecto de una solución ideal y a la mayor distancia respecto a una solución anti-ideal.
- El procedimiento consta de una serie de pasos que permiten calcular un proporción llamado “ratio de similaridad” que valora el desempeño de cada alternativa.
- Es compensatorio.
- Se sugiere normalizar con el módulo del vector.
- Muy intuitivo y transparente.
- Si bien acepta múltiples medidas de distancia suele utilizarse la euclídea.
- Tolera los criterios de minimización.



## TOPSIS - SKCriteria

Como pueden realizarse diferentes experimentos variando las funciones de agregación o normalizaciones; skcriteria proevee el concepto de *pipelines*.

```python
from skcriteria.preprocessing.scalers import VectorScaler
from skcriteria.pipeline import mkpipe
from skcriteria.madm.similarity import TOPSIS 
```

```python
pipe = mkpipe(
    MinimizeToMaximize(), 
    SumScaler(target="weights"), 
    VectorScaler(target="matrix"), 
    TOPSIS()
)
pipe
```

## TOPSIS - SKCriteria

```python
result = pipe.evaluate(dm)
result
```

```python
result.e_.ideal, result.e_.anti_ideal
```

```python
pipe.transform(dm)
```

## ÉLimination et Choix Traduisant la REalité (ELECTRE) - SKCriteria

- Es una familia de métodos de la "escuela francesa".
- Son métodos no compensarios.
- Diseñan internamente indices de superación entre las alternativas.
- En particular ELECTRE1, no resuelve un ranking sino un kernel.


```python
from skcriteria.madm.electre import ELECTRE1

pipe = mkpipe(
    MinimizeToMaximize(), 
    SumScaler(target="weights"), 
    VectorScaler(target="matrix"), 
    ELECTRE1()
)
```
