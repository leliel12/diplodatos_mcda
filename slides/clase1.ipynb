## MCDA: El proceso completo

1. Identificar y definir las **alternativas** de decisin, que deben ser factibles, exhaustivas y excluyentes. 
2. Definir los **atributos a optimizar**. Identificar objetivos que son de maximizacin o de minimizacin.    
3. Asociar a cada objetivo o atributo, una **funcin de utilidad** que permita expresar una valoracin cuantitativa de cada alternativa respecto al atributo ($atributo + funcin = criterio$). 
4. Consignar las valoraciones de cada alternativa respecto de cada criterio en una matriz de decisin.
5. Fija umbrales de satisfaccin para cada criterio, de manera tal que se eliminen las alternativas que en algn criterio no los superan y Prenalisis de dominancia. 
6. Si se requiere, transformar las valoraciones de las alternativas correspondientes a criterios de maximizacin.
7. Ponderar o asignar valoraciones a los criterios, conforme a la importancia relativa que tienen para el decisor, mediante una funcin de utilidad. Consignar en la matriz de decisin los $U(C_j)= w_jv$
8. Escalar/Normalizar todas las valoraciones, consignadas en la matriz de decisin completa, para llevarlas a una escala comn y hacerlas comparables. Obtener as la matriz de decisin definitiva.
9. Seleccionar un **MTODO DE AGREGACIN**.

## Erapa 5: Prenalisis de dominancia y satisfaccin

- Fija umbrales de satisfaccin para cada criterio, de manera tal que se eliminen las alternativas que en algn criterio no los superan.
- **Prenalisis de dominancia**. 
    - Si existe una alternativa dominante es sin duda la mejor eleccin, si no se quiere de una ordenacin.
    - Una alternativa $A_i$ es dominada si existe al menos otra $A_j$, tal que $A_i$  peor o igual $A_j$ en 
      todos los criterios y estrictamente peor a $A_j$ para al  menos un criterio. 
    - En smbolos: $A_j \succeq A_i$
    
**Si aplicamos:** Solo se aceptaran alternativas que rindan al menos el $2\%$, y $FX$ es peor a todas las dems

## Etapa 5: `skcriteria.preprocessing.filters`

Siguiendo el unciado que rezaba: "Si aplicamos: Solo se aceptaran alternativas que rindan al menos el 2%, y  es peor a todas las dems"

Podemos utilizar el modulo `filters` para lograr ambos objetivos. 

En primer lugar podemos quedarnos con los $ROE \ge 2$.

La clase `filters.Filter` recibe como parmetros un *dict* en cual cada llave es el nombre de un criterio, y cada valor es una funcin arbitraria que recibe la columna criterio como parmetro y retorna un arreglo con la misma cantidad de elementos que el array, y slo conserva las alternativas donde todos los criterios son `True`

````python
from skcriteria.preprocessing import filters

flt = filters.Filter({"ROE": lambda crit: crit >= 2})
flt
````

````python
flt.transform(dm)
````

## Etapa 5: `skcriteria.preprocessing.filters`

Por simplicidad y rendimiento computacional, se ofrece una colleccion de 

- `filters.FilterEQ` `->` $=$ (igualdad)
- `filters.FilterGE` `->` $\ge$ (mayor o igual)
- `filters.FilterGT` `->` $\gt$ (mayor)
- `filters.FilterIn` `->` $\in$ (incluye)
- `filters.FilterLE` `->` $\le$ (menor o igual)
- `filters.FilterLT` `->` $\lt$ (menor)
- `filters.FilterNE` `->` $\neq$ (distinto)
- `filters.FilterNotIn` `->` $\notin$ (no esta incluido)

````python
dm = filters.FilterGE({"ROE": 2}, ignore_missing_criteria=True).transform(dm)
dm
````

## Etapa 5: `skcriteria.preprocessing.filters`

## Etapa 8: `skcriteria.preprocessing.scalers`

Todas las funciones de escalado (menos una que vamos a ver en la clase que sigue) viven en el mdulo `preprocessing.scalers`.

al igual que *PushNegatives*, todos estos transformadores operan sobre los pesos (`target="weights"`), o sobre la matriz de alternativas (`target="matrix"`), o sobre ambos (`target="both"`). 

Empecemos por el mas popular, el fraccin suma, `SumScaler` cuya transformacin esta dada por:

$$
    r_{ij} = \frac{a_{ij}}{\sum_{i=1}^{m}a_{ij}}
$$


````python
from skcriteria.preprocessing import scalers
skl = scalers.SumScaler(target="both")  # vamos a escalar la matriz y los pesos
skl
````

````python
sum_scaled = skl.transform(dm)
sum_scaled
````

## Etapa 8: `skcriteria.preprocessing.scalers`


La fraccin rango, es lo que se conoce como `MinMaxScaler`

$$
    r_{ij} = \frac{a_{ij} - \min{a_i}}{\max{a_i} - \min{a_i}}
$$

Esto hace que $r_{ij}$ en general pertezcan el intervalo $[0, 1]$. Este intervalo es configurable con el parametro `criteria_range`.


````python
skl = scalers.MinMaxScaler(target="both")
skl
````

````python
mm_scaled = skl.transform(dm)
mm_scaled                                                          
````

## Etapa 8: `skcriteria.preprocessing.scalers`


La fraccin del maximo o `MaxAbsScaler` se define como:

$$
    r_{ij} = \frac{a_{ij}}{\max{a_i}}
$$

````python
skl = scalers.MaxAbsScaler(target="both")
skl
````

````python
max_scaled = skl.transform(dm)
max_scaled
````

## Etapa 8: `skcriteria.preprocessing.scalers`

Mdulo del vector o `VectorScaler` esta definido como:

$$
r_{ij} = \frac{a_{ij}}{\sqrt{\sum_{i=1}^{m}a_{ij}^2}}
$$



````python
skl = scalers.VectorScaler(target="both")
skl
````

````python
vct_scaled = skl.transform(dm)
vct_scaled
````

## Etapa 8: `skcriteria.preprocessing.scalers`

No podria faltar el `StandarScaler`

$$
r_{ij} = \frac{a_{ij} - \mu({a_i})}{\sigma({a_i})}
$$



````python
skl = scalers.StandarScaler(target="both")
skl
````

````python
std_scaled = skl.transform(dm)
std_scaled
````

## Etapa 9: Mtodo Aditivo Simple (SAM) - Ponderacin Lineal - Suma Ponderada

- Es uno de los mtodos ms utilizados, por ser sencillo e intuitivo en su aplicacin. 
- *Es compensatorio.*
- Requiere que las evaluaciones y los pesos sean cardinales.
- Necesita una normalizacin previa de las evaluaciones
- Es muy sensible al procedimiento de normalizacin utilizado (Suele usarse `SumScaler`). 
- El supuesto terico ms fuerte es que requiere la existencia de una funcin de utilidad cardinal aditiva para los criterios.
- Falla si algun valor es $< 0$.
- Consiste en construir una funcin valor $U(A_i)$ para cada alternativa, de la forma:

$$U(A_i) = \sum_{j=1}^n w_j r_{ij}$$

- donde $w_j$ y $r_{ij}$, representan, respectivamente, a los pesos de los criterios
y las preferencias de cada alternativa con respecto a cada criterio, una
vez que han sido normalizados.

## Etapa 9: Mtodo Multiplicativo Simple (SMM) - Multiplicacin Ponderada

- Es lo mismo que suma ponderada pero con multiplicaciones. 
- *Es compensatorio.*
- Requiere que las evaluaciones y los pesos sean cardinales.
- Necesita una normalizacin previa de las evaluaciones.
- Los pesos se elevan en vez de multiplicarse.
- Suele potenciar valores extremos.
- Es muy sensible al procedimiento de normalizacin utilizado (Suele usarse `SumScaler`). 
- Falla si algn valor es $\le 0$.
- Consiste en construir una funcin valor $U(A_i)$ para cada alternativa, de la forma:

$$U(A_i) = \prod_{j=1}^n (r_{ij})^{w_j}$$

- donde $w_j$ y $r_{ij}$, representan, respectivamente, a los pesos de los criterios
y las preferencias de cada alternativa con respecto a cada criterio, una
vez que han sido normalizados.
