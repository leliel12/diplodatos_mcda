## Teora racional de la decisin 

Un problema de Decisin se presenta cuando, frente a un problema, existe ms de un curso de accin posible.

**Sujetos intervinientes:**
    
- **Decisor**: es quien tiene el poder y responsabilidad de tomar la decisin y responde por sus consecuencias.
- **Analista**: es el encargado de estructurar el problema y ayudar al decisor a visualizarlo.

**Elementos conocidos**

- Conjunto de alternativas:&nbsp;$X= \left\lbrace x_1, x_2,..., x_n\right\rbrace$
- Conjunto de estados de la naturaleza:&nbsp;$Y= \left\lbrace y_1, y_2,..., y_m\right\rbrace$
- Matriz de las compensaciones: 

$$\begin{bmatrix}
     C(x_1,y_1)&C(x_1,y_2) & ...  & C(x_1,y_m) \\
     C(x_2,y_1)&C(x_2,y_2) & ...  & C(x_2,y_m) \\
     ... & ... & ... & ... \\
     C(x_n,y_1)&C(x_n,y_2) & ...  & C(x_n,y_m) \\
    \end{bmatrix}$$


## Para decidir racionalmente se aplica un marco normativo justificado en:

- Las decisiones ptimas aumentan las posibilidades de supervivencia en un entorno competitivo.
- Axiomas proporcionan una explicacin aceptable del comportamiento de eleccin (*Atractivo intuitivo*).

## Propiedad 2: Transitividad

$A \succ B$ $\longrightarrow d(A)> d(B)$

Se da cuando una opcin es mejor que otra en un estado de la naturaleza y al menos tan buena en todos los dems estados.

## Propiedad 4: Invarianza 

Las diferentes representaciones  (marco) del mismo problema de eleccin deben producir la misma preferencia. Es decir, la preferencia entre las opciones debe ser independiente de su descripcin. 

|          | $y_1$ | $y_2$ | $$d(x) = E_y C(x,y)$$ |   |          | $y_1$ | $y_2$ | $$d(x) = E_y C(x,y)$$ |
|:--------:|:-----:|:-----:|:-------------------:|---|:--------:|:-----:|:-----:|:-------------------:|
|   $x_1$  |   5   |   8   |         6,8         |   |   $x_1$  |   8   |   5   |         6,8         |

| $P(y_i)$ | 0,40  | 0,60  |                     |   | $P(y_i)$ | 0,60  | 0,40  |                     |

## Problema del Modelo Normativo 

Las **desviaciones** del comportamiento real del modelo normativo son **demasiado amplias** como para ser ignoradas, **demasiado sistemticas** para ser descartadas como error aleatorio, y **demasiado fundamentales** para ser acomodadas relajando el sistema normativo.

Distintos tericos respondieron a las violaciones observadas de cancelacin y transitividad al debilitar la teora normativa a fin de mantener su estatus como modelo descriptivo. 

`Una teora de la decisin racional no puede proporcionar una descripcin adecuada del comportamiento de eleccin.`

## `skcriteria.preprocessing.weighters.StdWeighter`

- Utiliza el desvio estandar por criterio normalizado.
- Se altera en caso de invertir los criteriios ($1/Ci$) o en normalizaciones.

$$w_j = \frac{s_j}{\sum{s}}$$

Donde:
- $s_j$ Es el devio muestral de cada criterio $j$.
- $\sum{s}$ es la sum de todos los $s_j$.

```python
wght = weighters.StdWeighter()
wght
```

```python
wght.transform(dm)
```

## `skcriteria.preprocessing.weighters.EntropyWeighter`

- En este caso el criterio con mas "sorpresas" es el mas importante.
- Bastante robusto frente a ruidos y normalizaciones.

$$
    w_j = \frac{H_j}{\sum{H}}
$$

Donde:
- $H_j$ entropa del criterio $j$.
- $\sum{H}$ la suma de las entropas de todos los criterios.

```python
wght = weighters.EntropyWeighter()
wght
```

```python
wght.transform(dm)
```

## Technique for Order Preference by Similarity to Ideal Solution (TOPSIS)

- Se basa en el concepto de que es deseable que una alternativa determinada se ubique a la distancia ms corta respecto de una solucin ideal y a la mayor distancia respecto a una solucin anti-ideal.
- El procedimiento consta de una serie de pasos que permiten calcular un proporcin llamado ratio de similaridad que valora el desempeo de cada alternativa.
- Es compensatorio.
- Se sugiere normalizar con el mdulo del vector.
- Muy intuitivo y transparente.
- Si bien acepta mltiples medidas de distancia suele utilizarse la eucldea.
- Tolera los criterios de minimizacin.



## TOPSIS - SKCriteria

Como pueden realizarse diferentes experimentos variando las funciones de agregacin o normalizaciones; skcriteria proevee el concepto de *pipelines*.

```python
from skcriteria.preprocessing.scalers import VectorScaler
from skcriteria.pipeline import mkpipe
from skcriteria.madm.similarity import TOPSIS 
```
