## Teoría racional de la decisión 

Un problema de Decisión se presenta cuando, frente a un problema, existe más de un curso de acción posible.

**Sujetos intervinientes:**
    
- **Decisor**: es quien tiene el poder y responsabilidad de tomar la decisión y responde por sus consecuencias.
- **Analista**: es el encargado de estructurar el problema y ayudar al decisor a visualizarlo.



## Teoría racional de la decisión 

**Elementos conocidos**

- Conjunto de alternativas:&nbsp;$X= \left\lbrace x_1, x_2,..., x_n\right\rbrace$
- Conjunto de estados de la naturaleza:&nbsp;$Y= \left\lbrace y_1, y_2,..., y_m\right\rbrace$
- Matriz de las compensaciones: 

$$\begin{bmatrix}
     C(x_1,y_1)&C(x_1,y_2) & ...  & C(x_1,y_m) \\
     C(x_2,y_1)&C(x_2,y_2) & ...  & C(x_2,y_m) \\
     ... & ... & ... & ... \\
     C(x_n,y_1)&C(x_n,y_2) & ...  & C(x_n,y_m) \\
    \end{bmatrix}$$


## Para decidir racionalmente se aplica un marco normativo justificado en: 

- La eficacia de las personas para alcanzar sus objetivos (*Maximización*).
- Las decisiones óptimas aumentan las posibilidades de supervivencia en un entorno competitivo.
- Axiomas proporcionan una explicación aceptable del comportamiento de elección (*Atractivo intuitivo*). 





## Reglas del modelo normativo 
- Cancelación
- Transitividad
- Dominancia
- Invarianza

## Propiedad 2: Transitividad 

Las utilidades están representadas a través de las funciones de decisión d(x).

$A \succ B$ $\longrightarrow d(A)> d(B)$

|          | $y_1$ | $y_2$ | $$d(x) = E_y C(x,y)$$ |
|:--------:|:-----:|:-----:|:-------------------:|
|   $x_1$  |   8   |   2   |         3,2         |
|   $x_2$  |   4   |   6   |         5,6         |
| $P(y_i)$ | 0,20  | 0,80  |                     |


 En este caso $d(x_1)< d(x_2)$
 $x_2$ es preferible a $X_1$


## Propiedad 3: Dominancia

Se da cuando una opción es mejor que otra en un estado de la naturaleza y al menos tan buena en todos los demás estados.

|       | $$y_1$$ | $$y_2$$ | $$y_3$$ |   |       | $$y_1$$ | $$y_2$$ | $$y_3$$ |
|:-----:|:-----:|:-----:|:-----:|   |:-----:|:-----:|:-----:|:-----:|
| $x_1$ |   5   |   8   |   7   |   | $x_1$ |   6   |   2   |   3   |
| $x_2$ |   3   |   2   |   1   |   | $x_2$ |   5   |   2   |   3   |


## Propiedad 4: Invarianza 

Las diferentes representaciones  (marco) del mismo problema de elección deben producir la misma preferencia. Es decir, la preferencia entre las opciones debe ser independiente de su descripción. 

|          | $y_1$ | $y_2$ | $$d(x) = E_y C(x,y)$$ |   |          | $y_1$ | $y_2$ | $$d(x) = E_y C(x,y)$$ |
|:--------:|:-----:|:-----:|:-------------------:|---|:--------:|:-----:|:-----:|:-------------------:|
|   $x_1$  |   5   |   8   |         6,8         |   |   $x_1$  |   8   |   5   |         6,8         |
|   $x_2$  |   6   |   2   |         3,6         |   |   $x_2$  |   2   |   6   |         3,6         |
| $P(y_i)$ | 0,40  | 0,60  |                     |   | $P(y_i)$ | 0,60  | 0,40  |                     |

## Problema del Modelo Normativo 

Las **desviaciones** del comportamiento real del modelo normativo son **demasiado amplias** como para ser ignoradas, **demasiado sistemáticas** para ser descartadas como error aleatorio, y **demasiado fundamentales** para ser acomodadas relajando el sistema normativo.

Distintos teóricos respondieron a las violaciones observadas de cancelación y transitividad al debilitar la teoría normativa a fin de mantener su estatus como modelo descriptivo. 

`Una teoría de la decisión racional no puede proporcionar una descripción adecuada del comportamiento de elección.`

## Encuadre y evaluación de los resultados


**Elementos de la decisión:**

Rinde de las perspectivas depende de x,y (antes llamamos compensaciones)

Probabilidades asociadas al rinde de cada perspectiva = q (probabilidad de pérdida) , p (probabilidad de ganancias)

Para cada alternativa X habrá Valores V(y), definidos en ganancias y pérdidas con pesos de decisiones $\pi$

$$\begin{matrix}
Valor &total  & de & la & perspectiva & = \pi(p). V(y_1) +  \pi(q).V(y_2)
\end{matrix}   $$

## Encuadre y evaluación de los resultados




La respuesta a las pérdidas es más extrema que la respuesta a las ganancias. La renuencia común a aceptar una apuesta justa al lanzar una moneda sugiere que el desagrado de perder una suma de dinero excede el placer de ganar la misma cantidad. La función de valor propuesta es **cóncava para las ganancias y convexa para las pérdidas**, y 
**más pronunciada para las pérdidas que para las ganancias**.

## Encuadre y ponderación de eventos de azar. 

En teoría de la perspectiva, el valor de un resultado incierto se multiplica por un peso de decisión $\pi$(p) que, es una función monótona de p pero no es una probabilidad. 



La función de ponderación  tiene las siguientes propiedades: 

1- los eventos imposibles son descartados, es decir, $\pi$(0 ) = 0, y la escala se normaliza de modo que $\pi$(1 )= 1, pero la función no se comporta bien cerca de los puntos finales. 

2- para probabilidades bajas, $\pi$(p) > p, pero $\pi$(1 - p) <= 1 (subcertidumbre). Por lo tanto, las probabilidades bajas están sobredimensionadas, las probabilidades moderadas y altas son infraponderadas, y el último efecto es más pronunciado que el primero. 

3- $\pi$(p.r)/$\pi$(p) <= $\pi$(p.q.r)/$\pi$(p.q) para todos 0 < p, q, r <=1 (subproporcionalidad). Es decir, para cualquier razón de probabilidad fija r, la relación de ponderaciones de decisión está más cerca de la unidad cuando las probabilidades son bajas que cuando son altas.

## Sesgo

Un sesgo cognitivo es una discrepancia sistemática entre la respuesta "correcta" en una tarea de juicio, dada por una regla normativa formal, y la respuesta la respuesta real del decisor o del experto a dicha tarea.

- Definimos los sesgos motivacionales como aquellos en los que los juicios están influidos por la deseabilidad o indeseabilidad de los acontecimientos, las consecuencias, los resultados o las elecciones. *Ej: presupuestos optimistas en proyectos en los que soy el creador, subestimar el tiempo de un proyecto de software porque deseo esté rápido*

### Sesgo de Anclaje y debiasing

El sesgo se produce cuando la estimación de un valor numérico se basa en un valor inicial (ancla), que luego se no se ajusta lo suficiente para proporcionar la respuesta final

**Evidencia:** Tareas de esimación, deciciones de precios y negociaciones

**Estrategias** 

- Evitar anclas
- Proporcionar múltiples contraanclajes
- Usar diferentes expertos que usen diferentes anclas


### Sesgo de disponibilidad o facilidad de retirada  y debiasing

El sesgo se produce cuando la probabilidad de un evento que es fácil de recordar es exagerada

**Evidencia:** Estimaciones de frecuencia simple; frecuencia de eventos letales; eventos raros que están anclados en ejemplos recientes.

**Estrategias** 

- Realizar un entrenamiento de probabilidades
- Proporcionar contraejemplos
- Proporcionar estadísticas



### Sesgo de efecto de certidumbre  y debiasing

La gente prefiere las cosas seguras a las apuestas con utilidades esperadas similares; descuentan la utilidad de las cosas seguras de forma drástica cuando dejan de ser seguras

**Evidencia:** Probabilidad vs certidumbre; métodos equivalentes producen diferentes resultados. 

**Estrategias** 

- Evitar las cosas seguras en la explicitación de la utilidad
- Separar valor y explicitación de la utilidad
- Explorar la actitud de riesgo relativo paramétricamente

### Sesgo de igualación y debiasing

Este sesgo se produce cuando los responsables de la toma de decisiones asignan pesos similares a todos los objetivos (17) o probabilidades similares probabilidades similares a todos los eventos


**Evidencia:** Explicitación de probabilidades en árboles de decisión y elicitación de pesos en los árboles de valores.

**Estrategias**

- Clasificar los eventos u objetivos primero, y luego asignar la proporción ponderaciones
- Obtener ponderaciones o probabilidades jerárquicamente

### Sesgo de presentar en términos de ganancia o pérdida y debiasing

Este sesgo se produce cuando las descripciones alternativas de una elección y sus resultados (44), ya sea como ganancias o como pérdidas, y puede llevar a respuestas diferentes.

**Evidencia:** Varias áreas que implican la elección de opciones arriesgadas, la evaluación de una única opción sobre un atributo, y la forma en que se consecuencias se describen para promover una elección

**Estrategias** 

- Identificar claramente el status quo (SQ)
- Para las funciones de valor, exprese valores como cambios marginales del SQ
- Para las funciones de utilidad, elija  por separado las utilidades para las ganancias y las pérdidas.

### Sesgo de  representación del problema miope y debiasing

Este sesgo se produce cuando se adopta una representación basada en un modelo mental incompleto del problema de decisión

**Evidencia:** Focalización en un pequeño número de alternativas, un bajo número de objetivos, o un sólo estado de la naturaleza  

**Estrategias**

- Animar explícitamente a pensar sobre más objetivos, nuevas alternativas y otros posibles estados del futuro

### Sesgo de omisión de variables importantes  y debiasing

El sesgo se produce cuando una variable importante se se pasa por alto

**Evidencia:** Definición de objetivos; identificación de alternativas de decisión; y generación de hipótesis

**Estrategias**

- Pregunte por las alternativas y los objetivos
- Pregunte por situaciones extremas o inusuales
- Utilizar técnicas de elicitación en grupo

### Sesgo del proxy  y debiasing

 Los atributos proxy reciben mayores pesos que los respectivos objetivos fundamentales.
 
**Evidencia:**  La elicitación de las ponderaciones en utilidad y valor multiatributo. 

**Estrategias**

- Evitar los atributos proxy
- Construir modelos que relacionen los atributos con los objetivos fundamentales y proporcionar ponderaciones para objetivos fundamentales


### Sesgo de Insensibilidad al rango  y debiasing

Las ponderaciones de los objetivos no se ajustan adecuadamente a los cambios en el rango de atributos.


**Evidencia:** Elicitacin de pesos en la medida de valor de la utilidad multriatributo.

**Estrategias**

- Hacer explícitos los rangos de atributos y utilizar procedimientos de ponderación de los mismos
- Utilizar procedimientos de compensación o de fijación de precios
- Utilizar procedimientos de elicitación múltiples y comprobaciones cruzadas

### Sesgo de Escala y debiasing

Familia de sesgos estímulo-respuesta que comprende: sesgo de contracción, sesgo logarítmico de respuesta, sesgo de igualación de rango, sesgo de centrado, y el sesgo de igualdad de frecuencia. 

**Evidencia:** EValoración de medidas físicas y sociales de diversa índole.

**Estrategias**

- Desarrollar escalas que coincidan con los estímulos y las respuestas, siendo conscientes de estos sesgos
- Elegir las técnicas de escalado técnicas apropiadas para la tarea tarea en cuestión


### Sesgo de División  y debiasing

Este sesgo se produce cuando la forma los objetivos se agrupan en un árbol de valores afecta a sus pesos; o la forma de podar un árbol afecta a las probabilidades colocadas en las ramas restantes.

**Evidencia:**  Elicitación de pesos en modelos multicriterio. Obtención de probabilidades en árboles de decisión.

**Estrategias**

- Evitar divisiones con grandes ratios de probabilidad o peso
- Utilizar la estimación jerárquica de los pesos o las probabilidades
- Utilizar juicios de proporción en lugar de de la estimación directa o distribución de puntos




### Sesgo de Influencia del afecto y debiasing

Este sesgo se produce cuando existe una predisposición emocional a favor o en contra de un resultado resultado u opción específica que contamina los juicios.

**Evidencia:**  Varios estudios que evalúan el papel del afecto que provoca una relación percibida inversa entre las consecuencias positivas y negativas relacionadas con las pandemias y los peligros causados por el hombre, etc.  También hay pruebas de que el afecto influye en la estimación de las probabilidades de los acontecimientos.

**Estrategias**

- Evitar las descripciones cargadas de consecuencias en los atributos
- Comprobar los juicios con con protocolos alternativos.
- Utilizar varios expertos con puntos de vista alternativos.


### Sesgo de Confirmación y debiasing

El sesgo se produce cuando hay un deseo de confirmar la propia creencia, lo que lleva a una inconsciente selectividad en la adquisición y uso de las pruebas. 


**Evidencia:**  Varios escenarios experimentales, como en la obtención de información
tareas de selección, actualización de pruebas y evaluación de juicios propios. (79,80) También en contextos del mundo real, como el diagnóstico médico, el razonamiento judicial y el pensamiento científico

**Estrategias**

- Utilizar múltiples expertos con diferentes puntos de vista sobre las hipótesis
- Desafíe las evaluaciones de probabilidad con contrafactuales
- Buscar pruebas de hipótesis alternativas

### Sesgo de deseabilidad de un evento o consecuencia positiva y debiasing

El sesgo se produce cuando la deseabilidad de un resultado conduce a un aumento de la medida en que se espera que ocurra que se produzca. También se denomina "pensamiento deseoso" o "sesgo de optimismo

**Evidencia:**  Predicción de resultados en juegos de azar; impacto en las estimaciones de probabilidades de resultados futuros en previsión de expertos; estimaciones de costos y duración en proyectos; así como así como algún posible efecto en torneos deportivos.

**Estrategias**

- Utilizar múltiples expertos con puntos de vista alternativos
- Utilizar la regla de puntuación y realizar apuestas hipotéticas contra el evento o consecuencia deseada consecuencia deseada
- Utilizar la descomposición y evaluación realista de probabilidades parciales

### Sesgo de indeseabilidad de un evento negativo y debiasing

Este sesgo se produce cuando se quiere ser cauto  o conservador en las estimaciones que pueden estar relacionadas
con las consecuencias negativas.

**Evidencia:**  La mayoría de las pruebas relacionadas con las probabilidades de los acontecimientos vitales; pero también en estimaciones a largo plazo de acontecimientos futuros en la previsión de expertos y en estimaciones de riesgos y beneficios sobre tecnologías de riesgo; algunas evaluaciones de riesgo que están intencionadamente sesgadas hacia estimaciones "conservadoras" en cada paso.

**Estrategias**
- Utilizar múltiples expertos con puntos de vista alternativos
- Utilizar reglas de puntuación y realizar apuestas hipotéticas a favor de el evento o la consecuencia no deseada 
- Utilizar la descomposición y evaluación realista de probabilidades parciales para estimar el suceso probabilidad

### Sesgo de Conveniencia de opciones/elecciones y debiasing

Este sesgo lleva a sobrestimar o subestimar las probabilidades, consecuencias, valores, o ponderaciones en una dirección que favorece una alternativa deseada. 

**Evidencia:** estimaciones sesgadas de probabilidades e impactos en la evaluación de riesgos

**Estrategias**
- Utilizar el análisis con múltiples partes interesadas que aporten diferentes perspectivas de valor
- Utilizar múltiples expertos con diferentes opiniones
- Utilizar incentivos y niveles adecuados de responsabilidad

### Sesgo de Aversión a la ambigüedad/Paradoja de Ellsberg y debiasing

La gente tiende a preferir los juegos de azar con probabilidades explícitas sobre los juegos con con probabilidades difusas o no especificadas.

**Estrategias**

- Modelar y cuantificar la ambigüedad como una distribución de probabilidad
- Modelar como incertidumbre paramétrica (por ejemplo, sobre el parámetro de sesgo de un proceso de Bernoulii) o como distribución de probabilidad secundaria
